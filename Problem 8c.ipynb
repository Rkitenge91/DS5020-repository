```python
import numpy as np
import matplotlib.pyplot as plt

w = np.array([2, -1])
Xs = [np.array([3,4]), np.array([1,2]), np.array([2,4]),
      np.array([2,2]), np.array([4,8]), np.array([1,1])]
labels = [np.dot(w, x) for x in Xs]
classes = [1 if l>0 else (-1 if l<0 else 0) for l in labels]
colors = {1:'green', 0:'gray', -1:'red'}

plt.figure(figsize=(6,6))
for x, c in zip(Xs, classes):
    plt.scatter(x[0], x[1], color=colors[c], s=50)
    plt.text(x[0]+0.05, x[1]+0.05, f'{x.tolist()}')

# plot decision boundary: w.x = 0 -> x2 = 2 x1
xs = np.linspace(-1, 5, 100)
ys = 2*xs
plt.plot(xs, ys, 'k-', linewidth=1, label='w·x = 0')

# plot weight vector as arrow
origin = np.array([0,0])
plt.arrow(0, 0, w[0], w[1], head_width=0.15, length_includes_head=True)
plt.xlim(-1, 5.5)
plt.ylim(-2, 9)
plt.gca().set_aspect('equal', 'box')
plt.legend()
plt.title('Classifier points, decision boundary, and weight vector')
plt.show()
```
