```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([2,-1])
v = np.array([3,4])
lams = [0, 0.25, 0.5, 0.75, 1]
pts = [lam*u + (1-lam)*v for lam in lams]

xs = [p[0] for p in pts]
ys = [p[1] for p in pts]

plt.figure(figsize=(5,5))
plt.plot([u[0], v[0]], [u[1], v[1]], 'k--', label='line through u and v')
plt.scatter(xs, ys, c='red')
for i, lam in enumerate(lams):
    plt.text(xs[i]+0.05, ys[i]+0.05, f'λ={lam}')
plt.scatter([u[0], v[0]], [u[1], v[1]], c='blue')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.gca().set_aspect('equal', 'box')
plt.title('Convex combinations (λ in [0,1]) between u and v')
plt.show()
```
